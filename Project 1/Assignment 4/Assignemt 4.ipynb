### Consider the $L_{p}$ norm for p = 12 and plot the corresponding $R^{2}$ unit circle. Is this really a norm or not? Discuss why or why not!

```python
#4

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def getPoints(p):
    x1 = np.linspace(0,1,500)

    x2 = np.power(1 - np.power(x1,p),1/p)

    x2_neg = -x2.copy()
    x1 = np.concatenate((x1,-x1,x1,-x1))
    x2 = np.concatenate((x2,x2,-x2,-x2))
    return x1,x2
```

```python
pValues = [1/2,1,2,4,8]

for p in pValues:
    x1,x2 = getPoints(p)
    ax = sns.lineplot(x1,x2,palette=plt.cm.winter,estimator=None)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.gca().set_title('Unit Circle: p = ' + str(p))
    plt.show()
```
